```python
#dropping features that have high correlation (correlation >= 0.90)
df.drop(['name','Jitter:DDP','Shimmer:APQ3','MDVP:Shimmer(dB)','MDVP:Shimmer','MDVP:RAP','MDVP:PPQ','Shimmer:DDA','PPE','MDVP:APQ','MDVP:Jitter(Abs)'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
#splitting the features and label
X=df.drop(['status'],axis=1)
y=df.status
```

##### Dividing records in training and testing sets with test size 20% of whole dataset

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#building model using randomforest algorithm
model = RandomForestClassifier(n_estimators=42)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print("accuracy score: ", accuracy_score(y_test, y_pred))
print("Confusion Matrix: ", confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
